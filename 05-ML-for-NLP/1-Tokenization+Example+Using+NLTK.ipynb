**Goal**: Learn how to break text into meaningful units (sentences and words)

## Why Tokenization?
- ML models can't understand raw text
- We need to split text into smaller pieces (tokens)
- Tokens can be **sentences** or **words**
- This is the **first step** in any NLP pipeline

2. Word Tokenization (Text â†’ Words)

````python
##  Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
````

## ðŸ“Œ Step 1: Sentence Tokenization

**What it does**: Splits a paragraph into individual sentences

**How it works**: Detects sentence boundaries using:
- Period (.)
- Exclamation mark (!)
- Question mark (?)
- New lines

**Output**: A list of sentences (each sentence = 1 document)

````python
sent_tokenize(corpus)
````

````python
documents=sent_tokenize(corpus)
````

````python
type(documents)
````

````python
for sentence in documents:
    print(sentence)
````

````python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize
````

## ðŸ“Œ Step 2: Word Tokenization

**What it does**: Splits text into individual words + punctuation

**Important**: Punctuation marks become separate tokens!

Example:
```
"Hello, world!" â†’ ["Hello", ",", "world", "!"]
```

**Why it matters**: ML models work at the word level, not sentence level

````python
word_tokenize(corpus)
````

````python
for sentence in documents:
    print(word_tokenize(sentence))
````

````python
from nltk.tokenize import wordpunct_tokenize
````

````python
wordpunct_tokenize(corpus)
````

````python
from nltk.tokenize import TreebankWordTokenizer
````

````python
tokenizer=TreebankWordTokenizer()
````

````python
tokenizer.tokenize(corpus)
````
