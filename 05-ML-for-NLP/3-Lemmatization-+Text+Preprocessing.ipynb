## ðŸ”‘ Critical Concept: POS (Part of Speech)

Lemmatization **depends on POS tags**!

Example:
- `better` as adjective â†’ `good`
- `better` as adverb â†’ `well`

**Always specify the correct POS tag for best results!**

```python
## Q&A,chatbots,text summarization
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='v')
```

## POS Tags Quick Reference

| POS | Tag | Example |
|-----|-----|---------|
| Noun | 'n' | student, apple |
| Verb | 'v' | eating, go |
| Adjective | 'a' | good, beautiful |
| Adverb | 'r' | quickly, well |

**Default**: If no POS is specified, noun ('n') is assumed!

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```
